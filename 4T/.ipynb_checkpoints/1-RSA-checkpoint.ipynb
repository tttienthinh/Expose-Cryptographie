```python
import numpy as np
import math
```

```python
def pgcd(a, b):
    if a==0:
        return b
    else:
        return pgcd(b%a, a)
```

```python
def euclid(a, b):
    if a==0:
        return 0, 1
    else:
        x, y = euclid(b%a, a)
        return y-(x*(b//a)), x
```

```python
def n_phi_n(p, q):
    return p*q, (p-1)*(q-1)
```

```python
def calculate_e(w):
    e = np.random.randint(2, w)
    
    while pgcd(e, w)!=1:
        e = np.random.randint(2, w)
    return e
```

```python
def calculate_d(e, w):
    x, y = euclid(e, w)
    return x
```

```python
def keys(p, q):
    n, w = n_phi_n(p, q)
    e = calculate_e(w)
    d = calculate_d(e, w) % w
    return (
        (n, e), # Public key
        (n, d)  # Private key
    )
```

```python
(n, e), (_, d) = keys(23, 17)
n, e, d
```

```python
def encrypt(m, n, e):
    return (m**e) % n
```

```python
def decrypt(m, n, d):
    return (m**d) % n
```

```python
encrypt(12, n, e)
```

```python
decrypt(146, n, d)
```
