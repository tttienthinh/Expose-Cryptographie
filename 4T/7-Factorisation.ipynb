```python
import time
import matplotlib.pyplot as plt
import sys, json
```

```python
sys.setrecursionlimit(100_000)
```

```python
def facto(n):
    if n == 1:
        return [n]
    for x in range(2, int(n**0.5)):
        if n % x == 0:
            return [x] + facto(n//x)
    return [n]
```

```python
def timing(n):
    start = time.time()
    facto(n)
    return time.time() -start
```

```python
L = []
for i in range(1, 2_000_000_000):
    L.append(timing(i))
L
```
